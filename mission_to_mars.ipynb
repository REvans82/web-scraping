```python
# Dependencies
from bs4 import BeautifulSoup
import requests
import os
from splinter import Browser
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
```

```python
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
# NASA MARS NEWS
# url used for data scraping
url = "https://redplanetscience.com/"
# initiating the webdriver for instant Chrome browser
browser.visit(url)
# web page content
html = browser.html
# parsing html using BeautifulSoup
soup = BeautifulSoup(html, 'html.parser')
print(soup.prettify())
```

```python
soup
```

```python
results = soup.find_all("div", {"class": "list_text"})
```

```python
results
```

```python
# titles = [result.find_all("div", {"class": "content_title"}) for result in results]
```

```python
titles = soup.find("div", class_="content_title").text
para = soup.find("div", class_="article_teaser_body").text
print(f"Latest Title: {titles}")
print(f"Latest Paragraph: {para}")
```

```python
hemisphere_image_urls = []
img_list = browser.find_by_css("a.itemLink.product-item img")
for i in range(len(img_list)):
    full_dict = {}
    browser.find_by_css("a.itemLink.product-item img")[i].click()
    full_dict['title'] = browser.find_by_css("h2.title").text
    sample_button = browser.find_by_text("Sample")
    sample_button.click()
    full_dict['img_url'] = sample_button['href'] #browser.find_by_css('img')[0]['src']
    hemisphere_image_urls.append(full_dict)
    browser.back()
    
hemisphere_image_urls
```
